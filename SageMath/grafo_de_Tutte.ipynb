```python
# Importamos la funcin para crear el grafo de Tutte
from sage.graphs.graph_generators import GraphGenerators
G = GraphGenerators().TutteGraph()

# Configuramos la visualizacin del grafo
plot_options = {
    'vertex_size': 20,
    'vertex_color': 'gold',
    'vertex_labels': False,      # Desactivamos las etiquetas para una vista ms limpia
    'edge_color': 'dodgerblue',
    'edge_thickness': 1.5,
    'layout': 'spring',          # Usamos un diseo tipo "spring" para distribuir bien los vrtices
    'figsize': [10, 10]          # Ajustamos el tamao de la figura para que sea ms grande
}

# Dibujamos el grafo de Tutte embellecido
G.show(**plot_options)
```
