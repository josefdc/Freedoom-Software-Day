```python
# Importamos la función para crear el grafo de Tutte
from sage.graphs.graph_generators import GraphGenerators
G = GraphGenerators().TutteGraph()

# Configuramos la visualización del grafo
plot_options = {
    'vertex_size': 20,
    'vertex_color': 'gold',
    'vertex_labels': False,      # Desactivamos las etiquetas para una vista más limpia
    'edge_color': 'dodgerblue',
    'edge_thickness': 1.5,
    'layout': 'spring',          # Usamos un diseño tipo "spring" para distribuir bien los vértices
    'figsize': [10, 10]          # Ajustamos el tamaño de la figura para que sea más grande
}

# Dibujamos el grafo de Tutte embellecido
G.show(**plot_options)
```
