```python
# Parmetros de visualizacin
xmin, xmax, ymin, ymax = -1.5, 1.5, -1.5, 1.5
width, height = 400, 400

# Definicin de la funcin para el fractal de Julia
def julia(z, c, max_iter):
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Funcin para generar la matriz de pxeles del conjunto de Julia para un valor dado de c y nmero de iteraciones
def julia_matrix(c, max_iter=100):
    return matrix([[julia(complex(x,y), c, max_iter) for x in srange(xmin, xmax, (xmax-xmin)/width)] for y in srange(ymax, ymin, (ymin-ymax)/height)])

# Generar una lista de imgenes del conjunto de Julia para diferentes valores de c
cs = [complex(0.355, 0.355), complex(-0.70176, -0.3842), complex(-0.8, 0.156), complex(-0.7269, 0.1889)]
frames = [matrix_plot(julia_matrix(c), cmap='hot', origin='lower') for c in cs]

# Combinar las imgenes en una animacin
animation = animate(frames, figsize=[6,6], axes=False)

# Mostrar la animacin
animation.show()
```
