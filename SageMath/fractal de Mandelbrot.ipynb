```python
# Parámetros de visualización
xmin, xmax, ymin, ymax = -2, 0.8, -1.5, 1.5
width, height = 400, 400  # Menor resolución para acelerar la generación de la animación

# Definición de la función para el fractal de Mandelbrot
def mandelbrot(c,max_iter):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Función para generar la matriz de píxeles del conjunto de Mandelbrot para un número dado de iteraciones
def mandelbrot_matrix(max_iter):
    return matrix([[mandelbrot(complex(x,y), max_iter) for x in srange(xmin, xmax, (xmax-xmin)/width)] for y in srange(ymax, ymin, (ymin-ymax)/height)])

# Generar una lista de imágenes del conjunto de Mandelbrot para diferentes números de iteraciones máximas
frames = [matrix_plot(mandelbrot_matrix(i), cmap='hot', origin='lower') for i in range(5, 51, 5)]  # Desde 5 hasta 50 iteraciones, de 5 en 5

# Combinar las imágenes en una animación
animation = animate(frames, figsize=[6,6])

# Mostrar la animación
animation.show()
```
