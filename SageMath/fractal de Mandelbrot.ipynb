```python
# Parmetros de visualizacin
xmin, xmax, ymin, ymax = -2, 0.8, -1.5, 1.5
width, height = 400, 400  # Menor resolucin para acelerar la generacin de la animacin

# Definicin de la funcin para el fractal de Mandelbrot
def mandelbrot(c,max_iter):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Funcin para generar la matriz de pxeles del conjunto de Mandelbrot para un nmero dado de iteraciones
def mandelbrot_matrix(max_iter):
    return matrix([[mandelbrot(complex(x,y), max_iter) for x in srange(xmin, xmax, (xmax-xmin)/width)] for y in srange(ymax, ymin, (ymin-ymax)/height)])

# Generar una lista de imgenes del conjunto de Mandelbrot para diferentes nmeros de iteraciones mximas
frames = [matrix_plot(mandelbrot_matrix(i), cmap='hot', origin='lower') for i in range(5, 51, 5)]  # Desde 5 hasta 50 iteraciones, de 5 en 5

# Combinar las imgenes en una animacin
animation = animate(frames, figsize=[6,6])

# Mostrar la animacin
animation.show()
```
